```python
import matplotlib.pyplot as plt
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120]
```

```python
plt.hist(blood_sugar_men,color='g',bins=[80,100,125,150], rwidth=0.98)

```

```python
plt.title("Blood sugar analysis")
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.hist([blood_sugar_men, blood_sugar_women],color=['green', 'orange'],bins=[80,100,125,150], rwidth=0.98, label=["men", "women"], orientation="horizontal")
plt.legend()
```
