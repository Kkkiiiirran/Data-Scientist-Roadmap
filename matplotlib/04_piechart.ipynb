```python
import matplotlib.pyplot as plt 
```

```python
exp_values = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car ", "Other Utilities"]
```

```python
plt.pie(exp_values, labels=exp_labels)
```

Radius


```python
plt.pie(exp_values, labels=exp_labels, radius=1.2)
plt.show()
```

Decimal Places

```python
plt.pie(exp_values, labels=exp_labels, autopct='%0.1f%%f')
plt.show()
```

```python
plt.pie(exp_values, labels=exp_labels, explode=[0,0.2,0,0,0], shadow=True)
plt.show()
```

```python
plt.pie(exp_values, labels=exp_labels, startangle=140)
plt.show()
```
