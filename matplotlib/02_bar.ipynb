```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
companies = np.arange(len(company))
companies
```

```python
plt.xticks(companies, company)

plt.title("US Tech Stocks")
plt.ylabel("Revenue(Bln)")
plt.bar(x=companies-.2, height=revenue, width = .4, label="Revenue")
plt.bar(companies+0.2, profit,width=0.4, label="Profit")
plt.legend()
```
